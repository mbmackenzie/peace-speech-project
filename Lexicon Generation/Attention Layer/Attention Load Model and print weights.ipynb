```
dict_total_peaceful = {}
for i in range(len(df_peaceful)):
  sample_dict = df_peaceful['dictionary_eval'][i]
  for key, value in sample_dict.items():
    try:
      dict_total_peaceful[key].append(value)
    except:
      dict_total_peaceful[key] = [value]
```

```
dict_total_non_peaceful = {}
for i in range(len(df_non_peaceful)):
  sample_dict = df_non_peaceful['dictionary_eval'][i]
  for key, value in sample_dict.items():
    try:
      dict_total_non_peaceful[key].append(value)
    except:
      dict_total_non_peaceful[key] = [value]
```

```
dict_average_peaceful = {}
for key, value in dict_total_peaceful.items():
  dict_average_peaceful[key] = sum(value)/len(value) 
```

```
dict_average_non_peaceful = {}
for key, value in dict_total_non_peaceful.items():
  dict_average_non_peaceful[key] = sum(value)/len(value) 
```

```
dict_peaceful_rank = sorted(dict_average_peaceful.items(),key=lambda x: x[1], reverse = True)[:400]
```

```
peaceful_400 = [vocab for vocab, weight in dict_peaceful_rank]
```

```
dict_non_peaceful_rank = sorted(dict_average_non_peaceful.items(),key=lambda x: x[1], reverse = True)[:400]
non_peaceful_400 = [vocab for vocab, weight in dict_non_peaceful_rank]
```

```
len(peaceful_400), len(non_peaceful_400)
```

```
peaceful_400 = set(peaceful_400)
nonpeaceful_400 = set(non_peaceful_400)

peaceful_vocab = list(peaceful_400 - nonpeaceful_400)
nonpeaceful_vocab = list(nonpeaceful_400 - peaceful_400)
```
