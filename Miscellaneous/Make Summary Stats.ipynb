```python
import platform
import os
import pandas as pd
```

```python
counter = 0

df_dict = {"year": [], "month": [], "publisher": [], "country":[],"id":[], "path":[]}

for entry in os.scandir(base_path):
    if os.path.isdir(entry):
        for entry_2 in os.scandir(entry.path):
            if os.path.isdir(entry_2):
                for entry_3 in os.scandir(entry_2.path):
                    if os.path.isdir(entry_3):
                        for entry_4 in os.scandir(entry_3.path):
                            counter += 1
                            try:
                                with open(entry_4, "r") as f:
                                    file_name_extention = os.path.basename(entry_4.path)
                                    file_name = file_name_extention.split('.')[0]
                                    _, country, date = file_name.split('_')
                                    day, month, year = date.split('-')
                                    article_id = f.readline().strip()
                                    article_title = f.readline().strip()
                                    publisher = f.readline().strip()
                                    url = f.readline()
                                    f.readline()
                                    article_text = f.readline().strip()
                                    df_dict['id'].append(article_id)
                                    df_dict['publisher'].append(publisher)
                                    df_dict['path'].append(entry_4.path)
                                    df_dict['year'].append("20" + str(year))
                                    df_dict['month'].append(month)
                                    df_dict['country'].append(country)
#                                     print(entry_4)
#                                     print(year, month, publisher, country, article_id)
                                    
                            except:
                                pass
                            
#                             break
#                     break
#             break
#     break
                    

counter
```
